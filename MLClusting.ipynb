```
from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import ClusteringEvaluator
```

```
movie = spark.sql('select * from movie1')
```

```
# vector assembler
assembler = VectorAssembler(inputCols=['day', 'month', 'year', 'genre', 'director_number', 'actor_number', 'budget', 'popularity', 'revenue', 'runtime', 'company', 'country', 'language', 'award','vote_average'], outputCol='features')

movie1 = assembler.transform(movie)
```

```
kmeans = KMeans(k=5,seed=1).fit(movie1)
evaluator = ClusteringEvaluator()
```

```
transformed = kmeans.transform(movie1).select("features", "prediction")
evaluator.evaluate(transformed)
```
